<h2> <span style="color:green">Relationship Between Job Type, Subscription Status, and Average Account Balance</span></h2>


```python
pivot_table = pd.pivot_table(a, 
                              values='balance', 
                              index='job', 
                              columns='deposit', 
                              aggfunc='mean', 
                              fill_value=0).round(2)

pivot_table
```

<h1><span style="color:green">Housing Loan vs Deposit Subscription </span></h2>

```python
housing_ct = pd.crosstab(a['housing'], a['deposit'], normalize='index') * 100
housing_ct
```

### **Description** :*This reveals a strong pattern where clients without a housing loan are much more likely to subscribe to a term deposit.*

```python

housing_ct.plot(kind='bar', stacked=True, colormap='rainbow',edgecolor="black")
plt.title("Housing Loan vs Deposit Subscription")
plt.ylabel("Percentage")
plt.xlabel("Housing Loan")
plt.xticks(rotation=0)
plt.legend(title="Deposit")
plt.tight_layout()
plt.show()
```

<h1><span style="color:green"> Overall Distribution of Term Deposit Subscription</span></h2>

```python
deposit_counts = a['deposit'].value_counts()
colors = ['#ff9999','#66b3ff']
plt.pie(deposit_counts, labels=deposit_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Overall Distribution of Term Deposit Subscription")
plt.axis('equal') 
plt.show()
```

### **Description**: *This pie chart shows what percentage of clients subscribed (yes) vs did not subscribe (no) to the term deposit. It gives an immediate overview of campaign effectiveness.*

<h1><span style="color:green"> Poutcome vs Deposit </span></h2>

```python

a['poutcome'] = a['poutcome'].replace(['unknown'], pd.NA)

a['poutcome'] = a['poutcome'].fillna(a['poutcome'].mode()[0])

```

```python
sns.countplot(data=a, x='poutcome', hue='deposit', palette='Set2')
plt.title('Poutcome vs Deposit (Effect of Previous Campaign Outcome)')
plt.xlabel('Previous Campaign Outcome')
plt.ylabel('Count')
plt.xticks([0, 1, 2], ['Failure', 'Success', 'Other'])
plt.tight_layout()
plt.show()
```

 ### **Description**: *A successful prior campaign increases the chances of a deposit in the current one.*

<h1><span style="color:green"> Average Call Duration By Month</span></h2>

```python

month_order = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 
               'jul', 'aug', 'sep', 'oct', 'nov', 'dec']

# Group by month and calculate average call duration
monthly_duration = a.groupby('month')['duration'].mean().reindex(month_order)

# Plot line chart
plt.plot(monthly_duration.index, monthly_duration.values, marker='o', linestyle='-', color='teal')
plt.title("Average Call Duration by Month")
plt.xlabel("Month")
plt.ylabel("Average Duration (seconds)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

 ### **Description**: *This shows how the average call duration (in seconds) varies across different months (month). It can reveal seasonal patterns in client engagement.*
